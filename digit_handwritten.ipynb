## 1. The Dataset

The [MNIST](http://yann.lecun.com/exdb/mnist/) dataset is one of the most famous datasets in machine learning, often called the "Hello World" of deep learning.

- **60,000 training images** and **10,000 testing images** of handwritten digits (0–9)
- Each image is a $28 \times 28$ grayscale grid (784 pixels total)
- Pixel values range from 0 (black) to 255 (white)
- **Input:** $28 \times 28$ grayscale pixels
- **Target:** A label (0 through 9) for each image

### Why MNIST?
- It's small enough to train on a CPU in seconds
- Simple enough for beginners, yet useful for benchmarking
- Real-world data — actual handwritten digits collected from census workers and high school students

### What We'll Do
1. **Load & explore** the dataset
2. **Build** a simple neural network
3. **Train** the model to recognize digits
4. **Evaluate** accuracy on unseen test data
5. **Predict** on individual images

````python
import tensorflow as tf
import matplotlib.pyplot as plt

# Load the dataset directly from Keras
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to be between 0 and 1
x_train, x_test = x_train / 255.0, x_test / 255.0

print(f"Training set: {x_train.shape}")
print(f"Test set: {x_test.shape}")

# Display a sample image
plt.imshow(x_train[0], cmap='gray')
plt.title(f"Label: {y_train[0]}")
plt.show()
````

## 2. The Neural Network Anatomy

We are building a **Feed-Forward Neural Network** (also called a Multi-Layer Perceptron). Data flows in one direction: input → hidden layer → output.

### Architecture Diagram
```
Input (28×28 image)
       ↓
┌─────────────┐
│   Flatten    │  28×28 → 784 values
└─────────────┘
       ↓
┌─────────────┐
│ Dense (128)  │  784 → 128 (ReLU activation)
│  Hidden Layer│  Learns patterns: edges, curves, loops
└─────────────┘
       ↓
┌─────────────┐
│ Dense (10)   │  128 → 10 (Softmax activation)
│   Output     │  Probability for each digit 0–9
└─────────────┘
```

### Layer Details

| Layer | Shape | What it does |
|-------|-------|-------------|
| **Flatten** | 28×28 → 784 | Stretches the 2D image into a 1D list of pixel values |
| **Dense (128, ReLU)** | 784 → 128 | The "hidden brain" — learns to detect patterns (lines, curves, loops). **ReLU** keeps only positive signals, helping the network learn non-linear relationships |
| **Dense (10, Softmax)** | 128 → 10 | Converts final signals into probability scores for digits 0–9. All probabilities sum to 1.0 |

### Key Terms
- **Dense layer:** Every neuron connects to every neuron in the previous layer (fully connected)
- **ReLU (Rectified Linear Unit):** $f(x) = \max(0, x)$ — simple but effective activation function
- **Softmax:** Turns raw scores into probabilities — e.g., `[0.01, 0.02, 0.85, 0.05, ...]` means 85% confident it's a "2"
