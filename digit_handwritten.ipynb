# Intro to Deep Learning with MNIST

**Goal:** Understand how a computer "sees" handwriting and predicts the number using a Feed-Forward Neural Network.

---

## 1. The Dataset

The MNIST dataset consists of **60,000 training images** and **10,000 testing images** of handwritten digits (0-9). Each image is a $28 \times 28$ grayscale grid.

- **Input:** $28 \times 28$ grayscale pixels
- **Target:** A label (0 through 9) for each image

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Load the dataset directly from Keras
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to be between 0 and 1
x_train, x_test = x_train / 255.0, x_test / 255.0

print(f"Training set: {x_train.shape}")
print(f"Test set: {x_test.shape}")

# Display a sample image
plt.imshow(x_train[0], cmap='gray')
plt.title(f"Label: {y_train[0]}")
plt.show()
```

## 2. The Neural Network Anatomy

We are building a model that flattens the $28 \times 28$ image into a single list of $784$ values.

| Layer | What it does |
|-------|-------------|
| **Flatten** | Stretches the image into a $784$-pixel line |
| **Dense (128 units)** | The "hidden" brain that identifies patterns (lines, curves) |
| **Softmax Output** | Converts the final signals into probability scores for digits $0-9$ |

```python
# Define the model architecture
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
model.fit(x_train, y_train, epochs=5)
```

## 3. The Learning Process

During training, the model uses an **Optimizer** to minimize the **Loss** (the error between its guess and the actual label).

- **Loss Function:** Measures how "wrong" the model's prediction is
- **Epochs:** One full pass through the training data

```python
# Evaluate the model on test data
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"\nTest Accuracy: {test_acc*100:.2f}%")
```
